```python
from skimage import io

doquyen1 = skimage.img_as_float(skimage.io.imread("datasets/images/doquyen1.jpg")).astype(np.float32)

print('Cac gia tri bieu dieu RGB thuoc kieu float tu 0 den 1')

print('\n')
print('Shape cua hoa do quyen')
print(np.shape(doquyen1))

plt.figure(figsize=(7,7))
plt.title(label='Hoa do quyen')
plt.imshow(doquyen1)
```

```python
def crop_image_center(image, cropx, cropy):
        height, width, channel = np.shape(image)
        
        start_y = height // 2 - (cropy // 2)
        start_x = width // 2 - (cropx // 2)
        
        stop_x = start_x + cropx
        stop_y = start_y + cropy
        
        return image[start_y: stop_y, start_x: stop_x]
```

```python
doquyen1_center_crop = crop_image_center(doquyen1, 300, 300)
print(np.shape(doquyen1_center_crop))
```

```python
plt.figure(figsize=(7,7))
plt.title('Hoa do quyen crop center')
plt.imshow(doquyen1_center_crop)
```
